```python
import pandas as pd 
file_handler = open("C:/Users/User/Desktop/insurance.csv", "r")
data = pd.read_csv(file_handler, sep = ",") 
file_handler.close() 
sex = {'male':1,'female':0} 
smoker={'yes':1,'no':0}
data.sex = [sex[item] for item in data.sex] 
data.smoker=[smoker[item] for item in data.smoker]
data.head(15)
```

```python
import seaborn as sns
corr=data.corr()
sns.heatmap(corr,xticklabels=corr.columns,yticklabels=corr.columns,annot=True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X = pd.DataFrame(np.c_[data['smoker'], data['age'],data['bmi']], columns = ['smoker','age','bmi'])
Y = data['charges']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=5)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
lin_model = LinearRegression()
lin_model.fit(X_train, Y_train)
```

```python
from sklearn.metrics import r2_score
# model evaluation for training set
y_train_predict = lin_model.predict(X_train)
rmse = (np.sqrt(mean_squared_error(Y_train, y_train_predict)))
r2 = r2_score(Y_train, y_train_predict)

print("The model performance for training set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
print("\n")

# model evaluation for testing set
y_test_predict = lin_model.predict(X_test)
rmse = (np.sqrt(mean_squared_error(Y_test, y_test_predict)))
r2 = r2_score(Y_test, y_test_predict)

print("The model performance for testing set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
```
